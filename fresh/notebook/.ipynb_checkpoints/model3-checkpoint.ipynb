```python
import pandas as pd
import numpy as np
```

```python
item_df = pd.read_csv('../data/tianchi_fresh_comp_train_item.csv')
user_df = pd.read_csv('../data/tianchi_fresh_comp_train_user.csv')
```

```python
user_df.head()
```

```python
user_data1 = user_df[['item_id', 'item_category']]
print user_data1.shape
user_data1.head()
```

```python
# 去重
user_data1 = user_data1.drop_duplicates()
print user_data1.shape
user_data1.head()
```

```python
user_data1['item_id'] = user_data1['item_id'].astype('str')
```

```python
user_data1_gb = user_data1.groupby(['item_category'])['item_id'].agg(lambda x: '|'.join(x)).reset_index()
```

```python
user_data1_gb.head()
```

```python
user_data1_gb['item_number'] = user_data1_gb.item_id.apply(lambda x: len(x.split('|')))
user_data1_gb.head()
```

```python
item_data1 = item_df[['item_id', 'item_category']]
item_data1.head()
```

```python
item_data1['item_id'] = item_data1['item_id'].astype('str')
```

```python
item_data1_gb = item_data1.groupby(['item_category'])['item_id'].agg(lambda x: '|'.join(x)).reset_index()
item_data1_gb.head()
```

```python
item_data1_gb['item_number'] = item_data1_gb.item_id.apply(lambda x: len(x.split('|')))
item_data1_gb.head()
```

```python
print user_data1_gb.shape
print item_data1_gb.shape
```

```python
user_data2 = user_df[['user_id', 'item_id', 'item_category', 'behavior_type', 'time']]
user_data2.head()
```

```python
user_data2['time_day'] = user_data2.time.apply(lambda x: x.split()[0])
user_data2['time_hour'] = user_data2.time.apply(lambda x: x.split()[-1])
user_data2 = user_data2.drop('time', axis=1)
user_data2.head()
```

```python
print user_data2.shape
```

```python
# 去重
user_data2 = user_data2.drop_duplicates()
print user_data2.shape
```

```python
user_data2['time_hour'] = user_data2['time_hour'].astype('int')
user_data2['time_hour_type'] = 0
```

```python
user_data2['time_hour_type'] = user_data2.time_hour.apply(lambda x: 1 if 2<=x<6 else (2 if 6<=x<10 else (3 if 10<=x<14 else (4 if 14<=x<18 else (5 if 18<=x<22 else 6)))))
user_data2 = user_data2.drop('time_hour', axis=1)
user_data2.head()
```

```python
user_data2['time_hour_type'].value_counts()
```

```python
user_data2['behavior_type'] = user_data2['behavior_type'].astype('str')
user_data2['time_hour_type'] = user_data2['time_hour_type'].astype('str')
```

```python
# one-hot
behavior_type = pd.get_dummies(user_data2['behavior_type'], prefix='behavior_type')
time_hour_type = pd.get_dummies(user_data2['time_hour_type'], prefix='time_hour')
```

```python
user_data3 = pd.concat([user_data2[['time_day', 'user_id', 'item_id', 'item_category']], behavior_type, time_hour_type], axis=1)
user_data3.head()
```

```python
user_data4 = pd.merge(user_data3, user_data1_gb[['item_category', 'item_number']], on='item_category', how='left')
user_data4.head()
```

```python
user_data4['item_number'].isnull().value_counts()
```

```python
# item_number数据标准化
item_number_mean = user_data4['item_number'].mean()
item_number_std = user_data4['item_number'].std()
print item_number_mean
print item_number_std
```

```python
user_data4['item_number'] = (user_data4['item_number'] - item_number_mean) / item_number_std
user_data4.head()
```

```python
user_data5 = user_df[user_df['behavior_type'] == 4][['item_id', 'item_category']]
user_data5.head()
```

```python
print user_data5.shape
```

```python
# 去重
user_data5 = user_data5.drop_duplicates()
print user_data5.shape
```

```python
user_data5['item_id'] = user_data5['item_id'].astype('str')
```

```python
user_data5_gb = user_data5.groupby(['item_category'])['item_id'].agg(lambda x: '|'.join(x)).reset_index()
user_data5_gb.head()
```

```python
user_data5_gb['item_buy_number'] = user_data5_gb.item_id.apply(lambda x: len(x.split('|')))
user_data5_gb.head()
```

```python
user_data6 = pd.merge(user_data1_gb[['item_category', 'item_number']], user_data5_gb[['item_category', 'item_buy_number']], on='item_category', how='left').fillna(0)
user_data6.head()
```

```python
user_data6['buy_rate'] = user_data6['item_buy_number'] / user_data6['item_number']
user_data6.head()
```

```python
user_data7 = pd.merge(user_data4, user_data6[['item_category', 'buy_rate']], on='item_category', how='left')
user_data7.head()
```

```python
user_data7['buy_rate'].isnull().value_counts()
```

```python
# buy_rate标准化
buy_rate_mean = user_data7['buy_rate'].mean()
buy_rate_std = user_data7['buy_rate'].std()
print buy_rate_mean
print buy_rate_std
```

```python
user_data7['buy_rate'] = (user_data7['buy_rate'] - buy_rate_mean) / buy_rate_std
user_data7.head()
```

```python
user_data7 = user_data7.drop('item_category', axis=1)
```

* 用15和16号的特征作为训练集的特征，17号的购买行为作为训练集的分类标签
* 用16和17号的特征作为测试集的特征，18号的购买行为作为测试集的分类标签
* 用17和18号的特征预测19号的购买行为

```python
train_data_x = user_data7[(user_data7['time_day']=='2014-12-15') | (user_data7['time_day']=='2014-12-16')]
train_data_x.head()
```

```python
test_data_x = user_data7[(user_data7['time_day']=='2014-12-16') | (user_data7['time_day']=='2014-12-17')]
test_data_x.head()
```

```python
pred_data = user_data7[(user_data7['time_day']=='2014-12-17') | (user_data7['time_day']=='2014-12-18')]
pred_data.head()
```

```python
train_data_y = user_data7[user_data7['time_day']=='2014-12-17'][['user_id', 'item_id', 'behavior_type_4']]
train_data_y.head()
```

```python
test_data_y = user_data7[user_data7['time_day']=='2014-12-18'][['user_id', 'item_id', 'behavior_type_4']]
test_data_y.head()
```

```python
train_data = pd.merge(train_data_x, train_data_y, on=['user_id', 'item_id'], how='left').fillna(0)
train_data.head()
```

```python
test_data = pd.merge(test_data_x, test_data_y, on=['user_id', 'item_id'], how='left').fillna(0)
test_data.head()
```

```python
train_data.rename(columns={'behavior_type_4_x': 'behavior_type_4', 'behavior_type_4_y': 'label'}, inplace=True)
train_data.head()
```

```python
test_data.rename(columns={'behavior_type_4_x': 'behavior_type_4', 'behavior_type_4_y': 'label'}, inplace=True)
test_data.head()
```

```python
x_train = train_data.iloc[:, 3:15]
y_train = train_data.iloc[:, -1]
```

```python
x_test = test_data.iloc[:, 3:15]
y_test = test_data.iloc[:, -1]
```

```python
x_pred = pred_data.iloc[:, 3:15]
```

```python
y_train.value_counts()
```

正负样本比例不均，采用SMOTE算法进行过采样

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
over_sample = SMOTE()
x_train_over_sample, y_train_over_sample = over_sample.fit_sample(x_train, y_train)
```

```python
Counter(y_train_over_sample)
```

## Logistic Regression

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
clf = LogisticRegression()

precision = np.mean(cross_val_score(clf, x_train_over_sample, y_train_over_sample, cv=10, scoring='precision'))
recall = np.mean(cross_val_score(clf, x_train_over_sample, y_train_over_sample, cv=10, scoring='recall'))
F1 = np.mean(cross_val_score(clf, x_train_over_sample, y_train_over_sample, cv=10, scoring='f1'))

print precision
print recall
print F1
```

```python
# 测试
clf.fit(x_train_over_sample, y_train_over_sample)
y_pred = clf.predict(x_test)
```

```python
print metrics.precision_score(y_test, y_pred)
print metrics.recall_score(y_test, y_pred)
print metrics.f1_score(y_test, y_pred)
```

```python
predict = clf.predict(x_pred)
```

```python
pred_data['label'] = predict
submission = pred_data[pred_data['label']==1]
submission = submission[['user_id', 'item_id']]
submission = submission.drop_duplicates()
submission.head()
```

```python
submission.to_csv('tianchi_mobile_recommendation_predict.csv', index=False)
```
